```python
import pandas as pd

class DictionaryTranslator:
    def __init__(self, dictionary_df):
        self.dictionary_df = dictionary_df.set_index('informal')['formal'].to_dict()

    def translate_sentence(self, sentence):
        words = sentence.split()
        translation = ' '.join(self.dictionary_df.get(word, word) for word in words)
        return translation

# Contoh penggunaan
# Buat instance Translatorm
translator = DictionaryTranslator(Kamus)

# Contoh penggunaan
input_sentence = input()
translation = translator.translate_sentence(input_sentence)
print(f'Kata Informal : {input_sentence}')
print(f'Kata Formal: {translation}')
```

```python
from nltk.translate.bleu_score import sentence_bleu
import nltk

# Fungsi untuk mengubah kalimat ke dalam token
def tokenize_sentence(sentence):
    return nltk.word_tokenize(sentence.lower())

# Kata Formal (Referensi Manusia)
formal_reference_sentence = "banget"

# Tokenize kalimat
informal_tokens = tokenize_sentence(input_sentence)
formal_reference_tokens = tokenize_sentence(formal_reference_sentence)
formal_translated_tokens = tokenize_sentence(translation)

# Menghitung skor BLEU
bleu_score = sentence_bleu([formal_reference_tokens], formal_translated_tokens)

# Menampilkan hasil
print(f'Informal: {informal_sentence}')
print(f'Formal Reference: {formal_reference_sentence}')
print(f'Formal Translated: {formal_translated_sentence}')
print(f'BLEU Score: {bleu_score}')
```
