```python
# Inference
from transformers import pipeline
model_checkpoint = "aditnnda/machine_translation_informal2formal"
translator = pipeline("translation", model=model_checkpoint)
```

# Inputan Kalimat Indonesia Informal ke Formal dan Perhitungan skor BLEU

```python
#  Input New Sentences
informal = input('Masukkan Kalimat Indonesia Informal: ')
translete = translator(informal)

formal = translete[0]['translation_text']
print(f'Kalimat `{informal}` dalam bentuk Formal yaitu: {formal}')
```

```python
# Calculate BLEU Score
labels = input('Kalimat formal yang benar adalah: ')

import evaluate
metric = evaluate.load("sacrebleu")
predictions = [formal]
references = [[labels]]

score = metric.compute(predictions=predictions, references=references)
print(f"Skor BLEU: {score['score']:.4f}%")
```
